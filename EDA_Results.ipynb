### Correlation Between Features
- for detecting Redundancy and Multicollinearity
- 
![Feature Correlation Heatmap](data/visualize/Correlation_between_features.png)

SMA_15, SMA_30, EMA_15, EMA_30, Rolling_Median_15, Lag_1, Lag_3, Lag_5
These are all strongly positively correlated (deep red blocks), perhaps only keep a couple to reduce multicollinearity.

Bollinger_Middle_15, Bollinger_Upper_15, and Bollinger_Lower_15 are almost perfectly correlated. Only keep 1.

Strong correlations among MACD_Line, MACD_Signal, and MACD_Histogram. Only keep 1.

Open, High, Low, Close are all strongly correlated. Only use the Close data.
